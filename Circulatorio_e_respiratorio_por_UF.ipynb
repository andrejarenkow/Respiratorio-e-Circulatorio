## Tentando um KMeans

### Preparando os dados

```
X = dados_juntos.iloc[:, [2,4]].values
X
```

### Método Elbow para definir a quantidade de clusters

```
from sklearn.cluster import KMeans
wcss = []
 
for i in range(1, 21):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(X)
    print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 21), wcss, marker='o')
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```

Pelo gráfico, a definição de quantos clusters vão ser é pela mudança brusca na curvatura. Eu vou escolher 6 clusters, inicialmente

### Fit o Dataset no modelo

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6, random_state=0) # aqui que se coloca a quantidade de clustersm deixei o random_state travado para gerar sempre os mesmos grupos
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```
#classificando de acordo com os grupos
labels = kmeans.labels_
labels
```

```
#Botando tudo no mesmo dataset
dados_juntos['grupo']=labels
dados_juntos.head()
```

```
# Trocando os label pra fazer mais sentido
dados_juntos.grupo = dados_juntos.grupo.replace([0,1,2,3,4,5],[4,0,2,5,1,3])
dados_juntos.head()
```

```
dados_juntos.to_csv('expo2.csv',sep=';', encoding='utf-8')
```

* 0 - 4,
* 1 -0,
* 2 - 2,
* 3 - 5,
* 4 - 1,
* 5 - 3





```
#Pontos dos Centros dos Clusters
centros = kmeans.cluster_centers_
centros = pd.DataFrame({'x':centros[:,0],'y':centros[:,1]})
centros['grupo']=centros.index
centros['grupo']=centros.grupo.replace([0,1,2,3,4,5],[4,0,2,5,1,3])
centros
```

### Visualizando em um gráfico

```
plt.scatter(X[:, 0], X[:, 1], c=dados_juntos['grupo'], s=100,alpha = 0.7, cmap = 'Reds',linewidths = .5, edgecolors='black')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black', s=300, alpha=0.8, linewidths = 1.5,edgecolors='white');
for line in range(0,centros.shape[0]):
     plt.text(centros.x[line]-0.015, centros.y[line]-0.009, centros.grupo[line], horizontalalignment='left', size='medium', color='white', weight='semibold')
plt.title('Relação entre doenças respiratórias e circulatórias', weight='semibold',size='large' )
plt.xlabel('internação por doença respiratória por 1000 habitantes')
plt.ylabel('internação por doença circulatória por 1000 habitantes') 
plt.show()
```

Definindo grupos:

0. internação circulatória maior do que respiratória - alta
1.   baixa internação das duas
2.   internação circulatória maior do que respiratória - baixa
3. alta internação circulatória e respiratória
4. internação respiratória maior do que circulatória - baixa
5. internação respiratória maior do que circulatória - alta
